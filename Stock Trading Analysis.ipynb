```python
#Analyze the return above
summary["mean"]=summary["mean"]*252
summary["std"]=summary["std"]*np.sqrt(252)
summary
```

```python
summary.plot.scatter(x="std",y="mean",figsize=(12,8),s=50, fontsize=15)
#annotate/create for loop 
for i in summary.index:
    plt.annotate(i,xy=(summary.loc[i,"std"]+0.002,summary.loc[i,"mean"]+0.002),size=15)
plt.xlabel("Annual risk(std)",fontsize=15)
plt.ylabel("Annual return",fontsize=15)
plt.title("Risk/Return",fontsize=24)
plt.show()
```

META has a high risk, but a low annual return.

MSFT has a lower risk than AAPL, but a similar return.

# Cummaltive Returns, Drawdowns...

```python
apple=yf.download("AAPL")
```

```python
apple
```

```python
apple=apple.Close.to_frame()
#We just want close data and store to apple
```

```python
apple
```

```python
apple["daily_returns"]=np.log(apple.div(apple.shift(1)))
#Shows daily returns
```

```python
apple
```

```python
apple.dropna(inplace=True)
```

```python
apple
```

```python
apple.daily_returns.sum()
```

```python
np.exp(apple.daily_returns.sum())
#The above line tells us that if $1 was invested in 1980, it would be worth $1175.
```

```python
apple["cummulative_returns"]=apple.daily_returns.cumsum().apply(np.exp)
```

```python
apple
```

```python
#To plot
apple.cummulative_returns.plot(figsize=(12,8),title="Apple Buy & Hold $1",fontsize=12)
plt.show()
```

```python
apple.daily_returns.mean()*252
```

Above gives avg annual return of apple stock

```python
apple.daily_returns.std()*np.sqrt(252)
```

Above gives high STD but Returns are high on annual basis.

```python
#How to calculate drawdowns
apple["cummulative_max"]=apple.cummulative_returns.cummax()
```

```python
apple
```

```python
#To plot
apple[["cummulative_returns","cummulative_max"]].plot(figsize=(12,8),title="Apple Buy & Hold $1 & Cummulative Max",fontsize=12)
plt.show()
```

Above graph shows the highs that Apple has created.

```python
apple["drawdown"]=apple["cummulative_max"]-apple["cummulative_returns"]
```

```python
apple
```

A drawdown refers to how much an investment or trading account is down from the peak before it recovers back to the peak.

```python
apple.drawdown.max()
```

$444 is the max drawdown apple has had.

```python
apple.drawdown.idxmax()
#This shows the date in which the $444 drawdown occured
```

```python
apple.loc[(apple.index=="2023-01-05")]
```

```python
apple.loc[(apple.index<="2023-01-05")]
#Shows data for less than or equal to
```

```python
#To show drawdown %
apple["drawdown%"]=(apple["cummulative_max"]-apple["cummulative_returns"])/apple["cummulative_max"]
```

# SMA Strategy

```python
data=apple.Close.loc[(apple.index>="1991-01-01")]
```

```python
data=data.to_frame()
```

```python
data
```

```python
sma_short=50
sma_long=100
```

```python
data["sma_short"]=data.Close.rolling(sma_short).mean()
data["sma_long"]=data.Close.rolling(sma_long).mean()
#To show Short and Long SMA
```

```python
data
```

```python
data.plot(figsize=(12,8),title="AAPL - SMA{} | SMA{}".format(sma_short,sma_long),fontsize=12)
plt.show()
```

```python
data.loc["2016"].plot(figsize=(12,8),title="AAPL - SMA{} | SMA{}".format(sma_short,sma_long),fontsize=12)
plt.show()
#Shows SMA for 2016. Can change .loc[""] for any date.
```

```python
data.dropna(inplace=True)
```

```python
data
```

```python
data["position"]=np.where(data["sma_short"]>data["sma_long"],1,-1)
#Tells us if SMA_Short is higher than SMA_LONG than we go Long.
```

```python
data
```

```python
data.loc[:,["sma_short","sma_long","position"]].plot(figsize=(12,8),title="AAPL = SMA{} | SMA{}".format(sma_short,sma_long), fontsize=12,secondary_y="position")
plt.show()
```

# Above, wanted to show when SMA strategy went Short or Long with extremes. Now to use shorter timeframe

```python
data.loc["2016",["sma_short","sma_long","position"]].plot(figsize=(12,8),title="AAPL = SMA{} | SMA{}".format(sma_short,sma_long), fontsize=12,secondary_y="position")
plt.show()
```

# Above shows the year 2016. It shows when the strategy went Short/Long. Dictated by +1 or -1

Time to backtest strategy.

```python
data["returns_b&h"]=np.log(data.Close.div(data.Close.shift(1)))
```

```python
data
```

```python
data["strategy"]=data["returns_b&h"]*data.position.shift(1)
```

```python
data
```

```python
data.dropna(inplace=True)
```

```python
data[["returns_b&h","strategy"]].sum()
```

Strategy had 1.96

Returns of Buy & Hold had 5.93

Buy&Hold Strategy beats SMA Strategy

However...compare more

```python
data[["returns_b&h","strategy"]].sum().apply(np.exp) 
#What $1 invested would be with each strategy.
```

```python
data[["returns_b&h","strategy"]].std()*np.sqrt(252)
#Annual std
```

```python
data.Close.plot(figsize=(12,8))
```

Seems like we have bias to upside.

Create a strategy where we don't go short

# Strategy adjusted with long bias

```python
data["position2"]=np.where(data["sma_short"]>data["sma_long"],1,0)
```

```python
data["strategy2"]=data["returns_b&h"]*data.position2.shift(1)
```

```python
data
```

```python
data.dropna(inplace=True)
```

```python
data[["returns_b&h","strategy2"]].sum()
```

```python
data[["returns_b&h","strategy2"]].sum().apply(np.exp) 
#What $1 invested would be with each strategy.
```

```python
## Strategy 2 is better than strategy 1.
##
## Showing Annual return for each strategy.
data[["returns_b&h","strategy2"]].std()*np.sqrt(252)
```
